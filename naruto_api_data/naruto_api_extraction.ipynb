```python
import requests
import pandas as pd
```

```python
base_url = "https://dattebayo-api.onrender.com/characters"

urls = []
all_character_data = []

for page_num in range(1,73):
    url = base_url + f"?page={page_num}"
    urls.append(url)

for i in urls:
    response = requests.get(i)
    if response.status_code == 200:
        character_data = response.json()
        all_character_data.append(character_data)
    else:
        print(f"Error: {response.status_code}")
```

```python
data = all_character_data

names = []
ids = []
debut_manga = []
debut_anime = []
debut_novel = []
debut_movie = []
debut_game = []
debut_ova = []
jutsu = []
nature_type = []
personal_sex = []
personal_bloodType = []
personal_kekkeiGenkai = []
personal_tailedBeasts = []
personal_teams = []
personal_clan = []
personal_classification = []

for item in data:
    characters = item['characters']

    for character in characters:
        names.append(character.get("name", "N/A"))
        ids.append(character.get("id", "N/A"))
        debut = character.get("debut", {})
        debut_manga.append(debut.get("manga", "N/A"))
        debut_anime.append(debut.get("anime", "N/A"))
        debut_novel.append(debut.get("novel", "N/A"))
        debut_movie.append(debut.get("movie", "N/A"))
        debut_game.append(debut.get("game", "N/A"))
        debut_ova.append(debut.get("ova", "N/A"))
        jutsu.append(character.get("jutsu", "N/A"))
        nature_type.append(character.get("natureType", "N/A"))
    
        try:
            personal = character["personal"]

            personal_sex.append(personal.get("sex", "N/A"))
            personal_bloodType.append(personal.get("bloodType", "N/A"))
            personal_clan.append(personal.get("clan", "N/A"))
            personal_kekkeiGenkai.append(personal.get("kekkeiGenkai", "N/A"))
            personal_teams.append(personal.get("team", "N/A"))
            personal_tailedBeasts.append(personal.get("tailedBeast", "N/A"))
            personal_classification.append(personal.get("classification", "N/A"))

        except:
            personal_sex.append("N/A")
            personal_bloodType.append("N/A")
            personal_clan.append("N/A")
            personal_kekkeiGenkai.append("N/A")
            personal_teams.append("N/A")
            personal_tailedBeasts.append("N/A")
            personal_classification.append("N/A")
```

```python
# Relevant CHARACTER detail saved to characters.csv

df_characters = pd.DataFrame({
    "id": ids,
    "name": names,
    "debut_manga": debut_manga,
    "debut_anime": debut_anime,
    "debut_novel": debut_novel,
    "debut_movie": debut_movie,
    "debut_game": debut_game,
    "debut_ova": debut_ova,
    "sex": personal_sex,
    "blood_type": personal_bloodType,
    "clan": personal_clan,
})

df_characters
```

```python
df_tailed_beasts = pd.DataFrame({
    "character_id": ids,
    "tailed_beast": personal_tailedBeasts,
})

df_tailed_beasts['tailed_beast'] = df_tailed_beasts['tailed_beast'].apply(lambda x: x.split(", ") if x != "N/A" else [])

df_tailed_beasts['tailed_beast'] = df_tailed_beasts['tailed_beast'].apply(lambda beasts: beasts if beasts else [None])

df_tailed_beasts = df_tailed_beasts.explode('tailed_beast')

df_tailed_beasts
```

```python
import re

df_tailed_beasts['has_variant'] = df_tailed_beasts['tailed_beast'].apply(lambda x: "(Forms)" in str(x))

tailed_beast_name = ["Ten-Tails", "Chōmei", "Gyūki", "Isobu", "Kokuō", "Kurama", "Matatabi", "Saiken", "Shukaku", "Son Gokū", "Nine-Tails"]

def extract_names(s):
    if s is None or s == "N/A":
        return "None"
    pattern = r'\b(?:' + '|'.join(re.escape(name) for name in tailed_beast_name) + r')\b'
    matches = re.findall(pattern, s)

    return ", ".join(matches) if matches else ""

df_tailed_beasts['tailed_beast_name'] = df_tailed_beasts['tailed_beast'].apply(extract_names)

df_tailed_beasts['variant_name'] = df_tailed_beasts.apply(lambda row: row['tailed_beast'] if row['has_variant'] else "no variant", axis=1)

df_tailed_beasts
```

```python
df_tailed_beasts.drop(columns=['tailed_beast'])

new_column_order = ['character_id', 'tailed_beast_name', 'has_variant', 'variant_name']

df_tailed_beasts = df_tailed_beasts[new_column_order]

df_tailed_beasts['tailed_beast_name'].replace(to_replace="Nine-Tails", value="Kurama", inplace=True)

df_tailed_beasts
```

```python
df_tailed_beasts.to_csv('/home/tmich20/python/naruto_api/naruto_api_data/data/tailed_beast_character.csv', index=False)
```
