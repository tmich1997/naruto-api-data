```python
import requests
import pandas as pd
```

```python
base_url = "https://dattebayo-api.onrender.com/characters"

urls = []
all_character_data = []

for page_num in range(1,73):
    url = base_url + f"?page={page_num}"
    urls.append(url)

for i in urls:
    response = requests.get(i)
    if response.status_code == 200:
        character_data = response.json()
        all_character_data.append(character_data)
    else:
        print(f"Error: {response.status_code}")
```

```python
data = all_character_data

names = []
ids = []
debut_manga = []
debut_anime = []
debut_novel = []
debut_movie = []
debut_game = []
debut_ova = []
jutsu = []
nature_type = []
personal_sex = []
personal_bloodType = []
personal_kekkeiGenkai = []
personal_tailedBeasts = []
personal_teams = []
personal_clan = []
personal_classification = []

for item in data:
    characters = item['characters']

    for character in characters:
        names.append(character.get("name", "N/A"))
        ids.append(character.get("id", "N/A"))
        debut = character.get("debut", {})
        debut_manga.append(debut.get("manga", "N/A"))
        debut_anime.append(debut.get("anime", "N/A"))
        debut_novel.append(debut.get("novel", "N/A"))
        debut_movie.append(debut.get("movie", "N/A"))
        debut_game.append(debut.get("game", "N/A"))
        debut_ova.append(debut.get("ova", "N/A"))
        jutsu.append(character.get("jutsu", "N/A"))
        nature_type.append(character.get("natureType", "N/A"))
    
        try:
            personal = character["personal"]

            personal_sex.append(personal.get("sex", "N/A"))
            personal_bloodType.append(personal.get("bloodType", "N/A"))
            personal_clan.append(personal.get("clan", "N/A"))

        except:
            personal_sex.append("N/A")
            personal_bloodType.append("N/A")
            personal_clan.append("N/A")
```

```python
# Relevant CHARACTER detail saved to characters.csv

df_characters = pd.DataFrame({
    "id": ids,
    "name": names,
    "debut_manga": debut_manga,
    "debut_anime": debut_anime,
    "debut_novel": debut_novel,
    "debut_movie": debut_movie,
    "debut_game": debut_game,
    "debut_ova": debut_ova,
    "sex": personal_sex,
    "blood_type": personal_bloodType,
    "clan": personal_clan,
})

df_characters
```

```python
df_characters.to_csv('/home/tmich20/python/naruto_api/naruto_api_data/data/characters.csv', index=False)
```

```python
df_jutsus = pd.DataFrame({
    "character_id": ids,
    "jutsu": jutsu
})

max_jutsus = df_jutsus['jutsu'].apply(len).max()

for i in range(max_jutsus):
    df_jutsus[f"jutsu_{i+1}"] = df_jutsus['jutsu'].apply(lambda x: x[i] if isinstance(x, list) and len(x) > i else None)

df_jutsus.drop('jutsu', axis=1, inplace=True)

df_jutsus
```

```python
df_jutsus['id'] = range(1, len(df_jutsus) + 1)

cols = df_jutsus.columns.tolist()
cols = ['id'] + [col for col in cols if col != 'id']
df_jutsus = df_jutsus[cols]

df_jutsus
```

```python
df_jutsus.to_csv('/home/tmich20/python/naruto_api/naruto_api_data/data/jutsus.csv', index=False)
```
