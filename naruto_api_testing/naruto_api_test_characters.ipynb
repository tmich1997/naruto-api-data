```python
# This is a test code only
# I want to know how the tables can be linked and simply experimenting
# characters -> link to: clans, kekkei-genkai, tailed beast, teams
# df_characters, df_jutsus, df_kekkeiGenkai, df_teams, df_clans, df_tailedBeasts, df_natureTypes

import requests
import pandas as pd
```

```python
url = "https://dattebayo-api.onrender.com/characters"

response = requests.get(url)

if response.status_code == 200:
    naruto_characters_data = response.json()
else:
    print(f"Error", response.status_code)
```

```python
characters = naruto_characters_data["characters"]

names = []
ids = []
manga_debut = []
anime_debut = []
novel_debut = []
movie_debut = []
game_debut = []
ova_debut = []
jutsus = []
nature_type = []
sex = []
kekkei_genkai = []
tailed_beasts = []
team = []
clan = []
```

```python
for character in characters:
    names.append(character.get("name", "N/A"))
    ids.append(character.get("id", "N/A"))
    debut = character.get("debut", {})
    manga_debut.append(debut.get("manga", "N/A"))
    anime_debut.append(debut.get("anime", "N/A"))
    novel_debut.append(debut.get("novel", "N/A"))
    movie_debut.append(debut.get("movie", "N/A"))
    game_debut.append(debut.get("game", "N/A"))
    ova_debut.append(debut.get("ova", "N/A"))
    jutsus.append(character.get("jutsu"))
    nature_type.append(character.get("natureType", "N/A"))
    personal = character.get("personal", {})
    sex.append(personal.get("sex", "unknown"))
    kekkei_genkai.append(personal.get("kekkeiGenkai", "N/A"))
    tailed_beasts.append(personal.get("tailedBeast", "N/A"))
    team.append(personal.get("team", "N/A"))
    clan.append(personal.get("clan", "N/A"))
```

```python
df_characters = pd.DataFrame({
    "id": ids,
    "name": names,
    "manga_debut": manga_debut,
    "anime_debut": anime_debut,
    "novel_debut": novel_debut,
    "movie_debut": movie_debut,
    "game_debut": game_debut,
    "ova_debut": ova_debut,
    "sex": sex,
    "tailed_beasts": tailed_beasts,
    "team": team,
    "clan": clan
})

df_characters
```

```python
df_jutsus = pd.DataFrame({
    "character_id": ids,
    "jutsus": jutsus
})

max_jutsus = df_jutsus['jutsus'].apply(len).max()

for i in range(max_jutsus):
    df_jutsus[f"jutsu_{i+1}"] = df_jutsus['jutsus'].apply(lambda x: x[i] if len(x) > i else None)

df_jutsus.drop('jutsus', axis=1, inplace=True)

df_jutsus
```

```python
# df_jutsus
# table COMPLETE

df_jutsus['id'] = range(1, len(df_jutsus) + 1)

cols = df_jutsus.columns.tolist()
cols = ['id'] + [col for col in cols if col != 'id']
df_jutsus = df_jutsus[cols]

df_jutsus
```

```python
# df_natureType
# table COMPLETE

df_natureType['id'] = range(1, len(df_natureType) + 1)

cols = df_natureType.columns.tolist()
cols = ['id'] + [col for col in cols if col != 'id']
df_natureType = df_natureType[cols]

df_natureType
```

```python
df_tailedBeast = pd.DataFrame({
    "character_id": ids,
    "tailed_beasts": tailed_beasts,
})

df_tailedBeast['tailed_beasts'] = df_tailedBeast['tailed_beasts'].apply(lambda x: x.split(", ") if isinstance(x, str) else [x])

max_tailedBeast = df_tailedBeast['tailed_beasts'].apply(len).max()

for i in range(max_tailedBeast):
    df_tailedBeast[f"tailedBeast_{i+1}"] = df_tailedBeast['tailed_beasts'].apply(lambda x: x[i] if len(x) > i else None)

df_tailedBeast.drop('tailed_beasts', axis = 1, inplace=True)

df_tailedBeast
```

```python
df_tailedBeast = pd.melt(df_tailedBeast, id_vars=['character_id'], value_vars=[f'tailedBeast_{i}' for i in range(1, 10)],
                  var_name='Tailed_Beast_Type', value_name='tailed_beast').dropna()

df_tailedBeast['tailed_beast'] = df_tailedBeast['tailed_beast'].replace("N/A", pd.NA)

df_tailedBeast = df_tailedBeast.dropna(subset=['tailed_beast'], axis=0)

df_tailedBeast.drop('Tailed_Beast_Type', axis = 1, inplace=True)



df_tailedBeast
```
